```python
x=[1,2,3,4,5,6,7]
y=[3,4,5,6,6.5,7,8]
x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3.5,3.7,4,4.5,5,5.2]
x2=[4,3,6,7,7.5,8,6.5,8,7.5,5,5.5]
y2=[4,4.5,5,6,5,4,4.5,4,6,5.5,4.5]
plt.scatter(x,y,label='high income low saving',color='r')
plt.scatter(x1,y1,label='low income high saving',color='b')
plt.scatter(x2,y2,label='loan',color='m')
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('scatter plot')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt                                      #this is called area plot
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='k',label='working',linewidth=5)
plt.plot([],[],color='b',label='playing',linewidth=5)
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','k','b'])
plt.xlabel('hours')
plt.ylabel('days')
plt.title('stackplot')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
branches=['cse','ece','it','ee']
exp = (0, 0, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
students_count=[15,30,25,10]
plt.pie(students_count,labels=branches,explode=exp,autopct='%1.1f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,
       colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown'])
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, hatch=['**O', 'oO', 'O.O', '.||.'])
```

```python
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%',
       pctdistance=1.25, labeldistance=.6)
```
