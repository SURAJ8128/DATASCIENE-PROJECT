```python
train=ImageDataGenerator(
      rescale=1/255,
      rotation_range=10,
      zoom_range=0.1,
      width_shift_range=0.1,
      height_shift_range=0.1)
```

```python
train.fit(x_train)
```

```python
train_generator=train.flow(x_train,y_train,batch_size=256)
```

```python
val_datagen=ImageDataGenerator(rescale=1/255)
```

```python
val_datagen.fit(x_test)
```

```python
val_Generator=val_datagen.flow(x_test,y_test)
```

```python
# create the model
model=tf.keras.models.Sequential([
      tf.keras.layers.Flatten(input_shape=(28,28,1)), #input layer
      tf.keras.layers.Dense(256,activation='relu'), #hL1
      tf.keras.layers.Dense(128,activation='relu'), #HL2
      tf.keras.layers.Dense(10,activation='softmax') #o/L
])
```

```python
from tensorflow.keras.optimizers import SGD

# Create an instance of the SGD optimizer
sgd_optimizer = SGD()

# Pass the instance to model.compile
model.compile(optimizer=sgd_optimizer,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(train_generator, epochs=20, verbose=1, validation_data=val_Generator)
```
